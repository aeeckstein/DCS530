12. **CreditCard**  Having or not having a credit card

1. **Personal_Loan**: Personal Loan: 0 = No personal loan given , 1 = personal loan given.

### Comparing Scenarios: Approved Loans VS. Declined

For comparison, we will split our data between loan request that were accepted VS. Declined

```python
df_accepted = bankLoanDF[bankLoanDF['Personal_Loan'] == 1]
df_declined = bankLoanDF[bankLoanDF['Personal_Loan'] == 0]
```

```python
high_Income_pmf = thinkstats2.Pmf(df_accepted['Income'], label = "Accepted")
low_Income_pmf = thinkstats2.Pmf(df_declined['Income'], label = "Declined")
```

```python
thinkplot.Pmfs([low_Income_pmf,high_Income_pmf])
thinkplot.Show(xlabel = 'Income', ylabel = 'PMF')
```

We can see, looking at our pmf's, the lower income the more likely it is to get declined for a loan.

### Computing A CDF

We will compute a CDF of income of accepted loans to visualize the aceptence rate

```python
income_amount =df_accepted.Income
income_cdf = thinkstats2.Cdf(income_amount, label = 'Income')
```

```python
thinkplot.Cdf(income_cdf)
thinkplot.Show(xlabel = 'Income', ylabel = 'CDF')
```

This figure shows us the cumulative probability of getting approved for a loan. As you can see the curve picks up at about 100K showing the income is a large factor in loan approval.

### Hypothesis Testing

The hypothesis I am aiming to test is whether there is a significant difference in Acceptance rate based of income.

```python
accepted_q = df_accepted.Income
declined_q = df_declined.Income
```

```python
from scipy.stats import ttest_ind

stat, p = ttest_ind(accepted_q, declined_q)
print('stat=%.3f, p=%.3f' % (stat, p))
```

The p-value for our hypothesis test is very close to 0, meaning that the difference in Acceptance rate based of income is statistically significant.
